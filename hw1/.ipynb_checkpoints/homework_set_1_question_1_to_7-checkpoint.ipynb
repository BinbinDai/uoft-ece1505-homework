Consider the function $f(x) = -\sum_{i=1}^{m}log(b_i - a_i^Tx)$, where $x \in R^n$, $b_i \in R$, $a_i \in R^n$. 

## Problem 1a
-----

Compute $\nabla f$ and $\nabla ^2 f$

## Solution
-------

$\nabla f = -\sum_{i=1}^{m} f_i(x)$

Let $f_i = log(b_i - a_i^Tx)$

$\nabla f_i = -a(b_i- a_i^Tx)^{-1}$

$\nabla f = \sum_{i=1}^{m} a(b_i- a_i^Tx)^{-1}$

$\nabla^2f = a_i a_i^T (b_i - a_i ^ T x)^{-2}$
where $a_i a_i^T \in R^{n\times n}$, $(b_i - a_i ^ T x)^{-2} \in R$ and $\nabla^2f \in R^{n\times n}$


## Problem 1b
Write down the first 3 terms of the taylor series expansion of $f(x)$ around some $x_0$ 

## Solution
----

$Taylor = f(x_o) + \nabla (f(x_0))^T (x - x_0) + \frac{(\nabla^2f(x))^T (x-x_0)^2}{2}$

$= -\sum_{i=1}^{m}log(b_i - a_i^Tx_o) + (\sum_{i=1}^{m} a(b_i- a_i^Tx_0)^{-1})^T (x - x_0) + (a_i a_i^T (b_i - a_i ^ T x_0)^{-2})^T  (x - x_0)^2$


# Question 2

What is the distance between two parallel hyperplanes 

$ HP_1 = \{x\in R^n \mid a^Tx=b_1 \} $ and 

$ HP_2 = \{x\in R^n \mid a^Tx=b_2 \} $

## Solution
$x \in R^n$ and $a \in R^n$

Let the line $L$ spanned by normal vector $a$ be $ L = \{x \mid at \}$ where $t \in R$ 

Let point $p_1$ be the point intersected by line $L$ and hyperplane $HP_1$ so $p_1 \in HP_1$ 

Let point $p_2$ be the point intersected by line $L$ and hyperplane $HP_2$ so $p_2 \in HP_2$

Then

$p_1 = a t_1$ and $a^Tp_1 = b_1 \implies$


\begin{align}
a^Tat_1 &= b_1\\
t_1 &= \frac{b_1}{\vert a \vert^2_2}
\end{align}

$p_2 = a t_2$ and $a^Tp_2 = b_2 \implies$


\begin{align}
a^Tat_2 &= b_2\\
t_2 &= \frac{b_2}{\vert a \vert^2_2}
\end{align}

$dist(p_1, p_2) = dist(HP_1, HP_2) = t_2 - t_1$

$\implies dist(HP_1, HP_2) = t_2 - t_1 = \frac{b_1-b_2}{\vert a \vert^2_2}$

```python
## Solution 7a/7b

f, J, H = Function('f'), Function('J'), Function('H')
x,y, b = symbols("x,y,beta")

func = x**2 + y**2 + b * x * y + x + 2*y
jacobian = Matrix([diff(func, x), diff(func, y)])
jac_solution = solve(jacobian, x, y)

hessian = Matrix([
    [diff(func, x, x), diff(func, x, y)],
    [diff(func, y, x), diff(func, y, y)]
])
```

### Problem 7b
For which values of $\beta$ is the $(x^*, y^*)$ a global minimum of $f(x,y)$ 

### Solution

If $f(x,y)$ is convex, when $\nabla f = 0$, $(x^*,y^*)$ is the global minimum

#### Prove that $f$ is convex

$f$ is convex if $\nabla^2 f$ is P.S.D

Let $\nabla^2 f = H$

{{L_hessian_equation}}


#### At what values of $\beta$ is H is P.S.D

**Approach 1: $R^{2\times 2} trick$**

With a 2 by 2 matrix

$H=\begin{bmatrix} a & b\\ c & d\end{bmatrix}$

Matrix is P.S.D
if $a \geq 0$ and $det \geq 0$

In this case, $a = 2 \geq 0$ and $det = 4 - \beta^2$ 

So $det \geq = 0$ when $-2 \leq \beta \leq 2$

```python
## approach 2 solution
v1, v2 = symbols("v_1,v_2")
v = Matrix([v1, v2])
h_equation = simplify((v.T * hessian * v)[0])

L_h_equation = get_latex(h_equation)
```

**Approach 2: H is P.S.D when $v^tHv \geq 0$**

$v^tHv=$ {{L_h_equation}}

Complete the square on $v_1$ gives us:

$ v_{1}^2 + \beta v_1v_2 + v_2^2 = (v_1+\beta v_2/2)^2 +\frac{4-\beta^2}{4}v_2^2 $

The whole equation is greater than 0 if $\frac{4-\beta^2}{4} \geq 0$

Which gives us

$-2 \leq \beta \leq 2$


**Visualization for Approach 2**

By subbing in some values for $\beta$, the equation is always positive when $-2 \leq \beta \leq 2$
